```python
import json 
import time

fp=open('record.json',"r")
data=fp.read()

products=json.loads(data)

prod_id=[]
prod_name = []
prod_price = []
prod_exp = []
prod_quan = []
sales={}
bill=0


print("********** Welcome to the Inventory Management Project ***********\n")
print("----------------------------------------------------------------")


while True:
  print("\nPress 1 if you want to add new inventory otherwise Press 2 if you want to Sale inventory \n")
  cho=int(input("Press here : "))
  while True:
    if cho == 1:
      
        new_prod_id = input("Enter the Product Id : ")
        new_prod_name = input("Enter the product Name : ")
        new_prod_price = int(input("Enter the product Price : "))
        new_prod_expdate = input("Enter the product ExpDate : ")
        new_prod_quantity = int(input("Enter the product Quantity : "))

        if new_prod_id in products.keys():
          products[new_prod_id]['Name'] = new_prod_name
          products[new_prod_id]['Price'] = new_prod_price
          products[new_prod_id]['ExpDate'] = new_prod_expdate
          products[new_prod_id]['Quantity'] = products[new_prod_id]['Quantity'] + new_prod_quantity

        else:
          products[new_prod_id] = {'Name' : new_prod_name, 'Price' : new_prod_price, 'ExpDate' : new_prod_expdate, 'Quantity' : new_prod_quantity}

        choose= input("\nDo you want to add more Inventories Y/N : ")
        print("----------------------------------------------------\n")
        if choose == 'Y':
          continue
        else:
          print(products)
          print("Your Inventory has Successfully added you can check into the record.json file")
          new_js = json.dumps(products)
          new_fd = open('record.json','w')
          new_fd.write(new_js)
          new_fd.close()
          break
        


    elif cho == 2:
      
      name=input("Please Enter the Customer  Name : ")
      phone=input("Please Enter the Customer Phone No.")
      print("\nPlease Select the product Code int the below list \n")
      for key, value in products.items():
        print(key, value['Name'])

      while True:
        prod=input("\nPlease Enter the product code which User want to Purchase : ")
        print("\n---------------- Selected Product Detail ---------------------\n")
        print("Name ",products[prod]['Name'])
        print("Price ",products[prod]['Price'])
        print("ExpDate ",products[prod]['ExpDate'])
        print("Quantity ",products[prod]['Quantity'])
        qun=int(input("Enter the Quantity user want to purchase : "))
        prod_id.append(prod)
        prod_name.append(products[prod]['Name'])
        prod_price.append(products[prod]['Price'])
        prod_exp.append(products[prod]['ExpDate'])
        prod_quan.append(qun)


        cont=input("\nDo you want to purchase more products Y/N : ")
        if cont == 'Y':
          continue
        else:
          break

      print("---------------------------------------------------------------------")
      print("\n\n******************* User has purchased these Products ************************")  
      for i in range(len(prod_price)):

        print("Name ",prod_name[i])
        print("Price ",prod_price[i])
        print("ExpDate ",prod_exp[i])
        print("Quantity ",prod_quan[i])
        print("----------------------------------------------------------------------")
        products[prod_id[i]]['Quantity']=products[prod_id[i]]['Quantity'] - prod_quan[i]


        bill+=prod_price[i]*prod_quan[i]
        sales[prod_id[i]]={"Name" : prod_name[i] , 'Price' : prod_price[i] ,'ExpDate' : prod_exp[i] ,
                          'Quantity' : prod_quan[i] , 'Time' : time.ctime()}

      sales['Total Bill']=bill
      cont_bill=input("\n Do you want to generate the bill Y/N : ")
      print("\n\n")
      if cont_bill == 'Y':
        print("Customer Name ",name)
        print("Phone No. ",phone)
        for i in prod_id:
          print("Name ",sales[i]['Name'])
          print("Price ",sales[i]['Price'])
          print("ExpDate ",sales[i]['ExpDate'])
          print("Quantity ",sales[i]['Quantity'])
          
          print("\n**********************************************************")
        print("Time ",sales[i]['Time'])
        print("Total Bill ",bill)

      js1=json.dumps(sales)
      fd1=open("Sales.json",'w')
      fd1.write(js1)
      fd1.close()

      js=json.dumps(products)
      fd = open("record.json",'w')
      fd.write(js)
      fd.close()
      break

    else:
      print("You have Press Wrong key Try Again ")
      break



```
